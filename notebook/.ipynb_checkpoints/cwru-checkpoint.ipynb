```python
def cum(series_data,delta_t=1.0/4000):
    print(delta_t)
    print(delta_t*len(series_data))
    
    lst_delta = []
    lst_cum = []
    cur =0
    delta=0
    cnt=0
    last=series_data[0]
    
    for i in series_data:
        delta = (i+last)/2* delta_t
        last=i
        cur = cur+delta
        
        lst_delta.append(delta)
        lst_cum.append(cur)
        cnt=cnt+1
    
    return pd.Series(lst_delta),pd.Series(lst_cum)
```
