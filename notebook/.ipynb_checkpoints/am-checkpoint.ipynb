```python
acc = pd.read_csv('acc.csv')
sr=4000
delta_t=1/sr
```

```python
def cum(series_data,delta_t=1/4000):
    lst_delta = []
    lst_cum = []
    cur =0
    delta=0
    for i in series_data:
        delta = i* delta_t
        cur = cur+delta
        lst_delta.append(delta)
        lst_cum.append(cur)
    
    return pd.Series(lst_delta),pd.Series(lst_cum)
```

```python
Delta_V_X,V_X=cum(acc['X'])
```
